###Identify Outliers

```python
def find_outliers_iqr(df, threshold = 1.5):
  outlier_summary = {}

  numeric_cols = df.select_dtypes("float64", "int64").columns

  for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    outlier_summary[col] = {
        "outlier_count": outliers.shape[0],
        "outlier_percentage" : 100 * outliers.shape[0] / df.shape[0],
        "lower_bound": lower_bound,
        "upper_bound": upper_bound

    }
  return pd.DataFrame(outlier_summary)
```

```python
find_outliers_iqr(df, threshold = 1.5)
```

```python
def remove_outliers_from_column(df,target_col, threshold = 1.5):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    return df[ (df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
def remove_outliers_from_all_columns(df, threshold = 1.5):
    df_clean = df.copy()
    numeric_cols = df.select_dtypes(include=["float64", "int64"]).columns

    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - threshold * IQR
        upper_bound = Q3 + threshold * IQR

        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean.copy()
```

```python
print("original data shape: ", df.shape)
df_target_clean = remove_outliers_from_column(df, "median_house_value")
print("only target column cleaning shape: ", df_target_clean.shape)
df_all_clean = remove_outliers_from_all_columns(df)
print("all columns cleaning shape: ", df_all_clean.shape)
```

###Transformations

```python
from sklearn.preprocessing import PowerTransformer

# Create transformed copies (do not overwrite originals)
X_train_transformed = X_train.copy()
X_test_transformed  = X_test.copy()
```

```python
# Columns to apply log1p (count-like, heavy right-skew)
log_cols = ["total_rooms", "total_bedrooms", "population", "households"]

# Keep only columns that exist in your dataframe
log_cols = [c for c in log_cols if c in X_train_transformed.columns]

# Apply log1p (safe for zeros)
for c in log_cols:
    X_train_transformed[c] = np.log1p(X_train_transformed[c])
    X_test_transformed[c]  = np.log1p(X_test_transformed[c])
```

```python
# Columns to apply Yeo-Johnson (usually mild skew columns)
yj_cols = ["median_income"]
yj_cols = [c for c in yj_cols if c in X_train_transformed.columns]

if len(yj_cols) > 0:
    pt = PowerTransformer(method="yeo-johnson", standardize=False)  # fit on train only
    X_train_transformed[yj_cols] = pt.fit_transform(X_train_transformed[yj_cols])
    X_test_transformed[yj_cols]  = pt.transform(X_test_transformed[yj_cols])
```

```python
use_log_target = True

if use_log_target:
    y_train_transformed = np.log1p(y_train)
    y_test_transformed  = np.log1p(y_test)

    # Inverse transform for predictions
    def inverse_target(y_pred_log):
        return np.expm1(y_pred_log)
else:
    y_train_transformed = y_train
    y_test_transformed  = y_test
    inverse_target = None
```

##Training

```python
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, classification_report, confusion_matrix, mean_absolute_error, mean_squared_error
```

```python
def evaluate_model(true, predicted):
  mae = mean_absolute_error(true, predicted)
  mse = mean_squared_error(true, predicted)
  rmse = np.sqrt(mean_squared_error(true, predicted))
  r2_square = r2_score(true, predicted)
  return mae, rmse, r2_square
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Ridge": Ridge(),
    "Lasso": Lasso(),
    "K-Neighbors Regressor": KNeighborsRegressor(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest Regressor": RandomForestRegressor(),
    "ADA Boost Regressor": AdaBoostRegressor(),
    "Gradient Boosting Regressor": GradientBoostingRegressor(),
    "XGBoost Regressor": XGBRegressor()
}
```

```python
for i in range(len(list(models))):
    model = list(models.values())[i]

    # Fit on transformed data
    model.fit(X_train_transformed, y_train_transformed)

    # Predict
    y_train_pred = model.predict(X_train_transformed)
    y_test_pred  = model.predict(X_test_transformed)

    # If target was log-transformed, bring predictions back to original scale
    if use_log_target:
        y_train_pred_eval = np.expm1(y_train_pred)
        y_test_pred_eval  = np.expm1(y_test_pred)

        # Evaluate against ORIGINAL y (not log)
        mae, rmse, r2_square = evaluate_model(y_train, y_train_pred_eval)
        model_test_mae, model_test_rmse, model_test_r2_square = evaluate_model(y_test, y_test_pred_eval)
    else:
        mae, rmse, r2_square = evaluate_model(y_train, y_train_pred)
        model_test_mae, model_test_rmse, model_test_r2_square = evaluate_model(y_test, y_test_pred)

    print(list(models.keys())[i])

    print("-------------------")
    print("Train MAE: ", mae)
    print("Train RMSE: ", rmse)
    print("Train R2 Score: ", r2_square)

    print("-------------------")
    print("Test MAE: ", model_test_mae)
    print("Test RMSE: ", model_test_rmse)
    print("Test R2 Score: ", model_test_r2_square)

    print("-------------------")
    print("\n")
```
